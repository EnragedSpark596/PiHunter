```python
# Python program to display all the prime numbers within an interval
# https://www.programiz.com/python-programming/examples/prime-number-intervals

lower = 6
upper = 30 #100

denom = []

for num in range(lower,upper + 1):
    if num > 1:
        for i in range(2,num):
            if (num % i) == 0:
                break
        else:
            denom.append(num)
print(denom)
```

```python
numer = [i*3+1 for i in denom]
```

```python
best_numerators=[]
errorList = []
for prime in denom:
    errorScores = []
    testNumRange = list(range(3*prime,4*prime))
    for num in testNumRange:
        # zero the error list for this iteration
        
        testResult = num/prime
        error = (math.pi - testResult)/math.pi
        errorScores.append(error)
        lowestError = min(errorScores)
        lowErrorIDX = errorScores.index(lowestError)
        bestNumerator = testNumRange[lowErrorIDX]
    errorList.append(errorScores)
    best_numerators.append(bestNumerator)
print(best_numerators)
```
